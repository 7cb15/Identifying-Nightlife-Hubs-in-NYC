```python
import pandas as pd
import geopandas as gpd
```

```python
#read in both pick up and drop off dataframes
pickup = pd.read_csv('pickup_vol.csv')
dropoff = pd.read_csv('dropoff_vol.csv')
```

```python
total = pd.merge(pickup,dropoff,how='outer',left_on='PULocationID',right_on='DOLocationID')
total = total[['PULocationID','night_counts_x','day_counts_x','night_prop_x','night_counts_y','day_counts_y','night_prop_y']]
total.fillna(0,inplace=True)
total.head()
```

```python
total.rename(columns={'night_counts_x':'PU_night','day_counts_x':'PU_day','night_prop_x':'PU_prop','night_counts_y':'DO_night','day_counts_y':'DO_day','night_prop_y':'DO_prop'},inplace=True)
```

```python
total.columns
```

```python
total['total_night'] = total['PU_night'] + total['DO_night'] 
total['total_day'] = total['PU_day'] + total['DO_day'] 
total['total_prop'] = total['total_night'] / (total['total_night'] + total['total_day'])
```

```python
taxi_geo = gpd.read_file('taxi_zones.shp')
```

```python
taxi_geo.head()
```

```python
pickup.head()
```

```python
pickup_geo = pd.merge(taxi_geo,pickup,how='left',left_on='LocationID',right_on='PULocationID')
pickup_geo.head()
```

```python
dropoff_geo = pd.merge(taxi_geo,dropoff,how='left',left_on='LocationID',right_on='DOLocationID')
dropoff_geo.head()
```

```python
total_geo = pd.merge(taxi_geo,total,how='left',left_on='LocationID',right_on='PULocationID')
```

```python
pickup_geo.to_file(driver='ESRI Shapefile',filename='pickup')
dropoff_geo.to_file(driver='ESRI Shapefile',filename='dropoff')
```

```python
total_geo.to_file(driver='ESRI Shapefile',filename='total')
```
